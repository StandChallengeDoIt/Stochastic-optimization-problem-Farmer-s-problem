# Homework 2: Farmer's problem resolution and simulation

### Definition of the Farmer's problem

```python
import random
import pulp

num_trials=5000

# Create a linear programming problem
def lp_problem(t1,t2,t3):
    prob = pulp.LpProblem("Farmers_Problem", pulp.LpMaximize)

    # Decision variables
    x1 = pulp.LpVariable("Acres_of_Wheat", lowBound=0)
    x2 = pulp.LpVariable("Acres_of_Corn", lowBound=0)
    x3 = pulp.LpVariable("Acres_of_Sugar_Beets", lowBound=0)
    w1 = pulp.LpVariable("Tons_of_Wheat_Sold", lowBound=0)
    y1 = pulp.LpVariable("Tons_of_Wheat_Purchased", lowBound=0)
    w2 = pulp.LpVariable("Tons_of_Corn_Sold", lowBound=0)
    y2 = pulp.LpVariable("Tons_of_Corn_Purchased", lowBound=0)
    w3 = pulp.LpVariable("Tons_of_Sugar_Beets_Sold", lowBound=0)
    w4 = pulp.LpVariable("Tons_of_Sugar_Beets_Sold", lowBound=0)
    # Objective function
    prob += 0

    # Constraints
    
    # Land constraint
    prob += x1 + x2 + x3 <= 500

    # Wheat feed requirement
    prob += t1*x1-w1 + y1 >= 200

    # Corn feed requirement
    prob += t2*x2-w2 + y2 >= 240

    # Sugar beet 
    prob += w3 +w4 <= t3*x3
    prob += w3  <= 6000

    prob.objective=(-(150*x1+230*x2+260*x3+238*y1-170*w1+210*y2-150*w2-36*w3-10*w4)
            )
    prob.solve()
    profit = pulp.value(prob.objective)
    return profit
```

### Perfect information solution

```python
# Below average
profit_perfect_information_below=lp_problem(2,2.4,16)
# Average average
profit_perfect_information_average=lp_problem(2.5,3,20)
# above average
profit_perfect_information_above=lp_problem(3,3.6,24)
Average_profit_perfect_information=(profit_perfect_information_below+profit_perfect_information_average+profit_perfect_information_above)/3
print("Average profit for perfect information: $", Average_profit_perfect_information)
```

### Stochastic solution 

```python
def stochastic_sim():
    
    total_profit_stochastic = 0
    for sim in range(num_trials):
    # First stage: the solutions are taken from the examples in the class x1,x2,x3
        x1=170
        x2=80
        x3=250
    # Second stage: The yield is randomly chosed between the 20% below or above average and compute  for the simulation
        a1=[2,2.5,3]
        a2=[2.4,3,3.6]
        a3=[16,20,24]
    # Random yield condition is considered
        b1= random.choice(a1)
        ind=a1.index(b1)
        b2 = a2[ind]
        b3 = a3[ind]
        problem2=lp_problem(b1,b2,b3)
        total_profit_stochastic += problem2
        print('simulation',sim,'total_profit_expected_value:$',total_profit_stochastic )
    return total_profit_stochastic / num_trials
#The simulation gives the True optimal value of the stochastic approach
profit_stochastic_solution=stochastic_sim()
print("Average profit for stochastic solution $", profit_stochastic_solution)                   
```

### Expected value solution 

```python
def expected_value_sim():
    
    total_profit_expected_value=0
    for sim in range(num_trials): 
    # Here the expected yield is at each iteration assumed of a certain (Expected value)
        a1=[2,2.5,3]
        a2=[2.4,3,3.6]
        a3=[16,20,24]
        b1= random.choice(a1)
        ind=a1.index(b1)
        b2 = a2[ind]
        b3 = a3[ind]
        #b1= random.choice(a1)
        #b2= random.choice(a2)
        #b3= random.choice(a3)
        #expected_yield=[b1,b2,b3]
        problem3=lp_problem(b1,b2,b3)
        total_profit_expected_value += problem3
        print('simulation',sim,'total_profit_expected_value:$',total_profit_expected_value)
    return total_profit_expected_value / num_trials
#The simulation gives the True average value of the expected value solution
total_profit_expected_value=expected_value_sim()
print("Average profit for expected value solution $", total_profit_expected_value)                   
```
